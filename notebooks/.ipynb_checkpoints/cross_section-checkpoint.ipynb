```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots()

# Create inverted triangle
x = np.array([-1, 0, 1])
y = np.array([5, 0, 5])

# Plot the inverted triangle
ax.fill(x, y, 'cyan', edgecolor='black', linewidth=3)

# Add axis lines
ax.axhline(0, color='black',linewidth=2)
ax.axvline(0, color='black',linewidth=2)

# Show the plot
plt.show()
```
