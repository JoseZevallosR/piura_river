```python
import numpy as np
from pySWATPlus.TxtinoutReader import TxtinoutReader
from pySWATPlus.SWATProblem import SWATProblem, minimize_pymoo
import random
import pandas as pd

from pymoo.algorithms.soo.nonconvex.cmaes import CMAES
from pymoo.termination import get_termination
from pymoo.util.normalization import denormalize
```

```python
"""
The user defined function must receive only one parameter, which is a dictionary taht can have any structure, but must have
one key (with any user-specified name, in our case is 'calibration_params') that contains the calibration parameters. It must be in the following format: 
{filename: (id_col, [(id, col, value)])}
"""

# Load historical flow data globally if possible to avoid reloading
# Assuming 'historical_data.csv' has columns 'date' and 'flow'
historical_flow_data = pd.read_csv('D:/Proyectos_GitHub/piura_river/data/historical_flow_data.csv')
historical_flow_data['date'] = pd.to_datetime(historical_flow_data['date'])

def calculate_error(simulated, observed):
    """
    Calculate error between observed and simulated flows.
    Here, we'll use the Nash-Sutcliffe Efficiency (NSE) as an example.
    """
    numerator = np.sum((observed - simulated) ** 2)
    denominator = np.sum((observed - np.mean(observed)) ** 2)
    return 1 - (numerator / denominator)

def function_to_minimize(dict_of_params):
    calibration_params = dict_of_params['calibration_params']
    path_to_txtinout = dict_of_params['path_to_txtinout']

    # Set up SWAT reader and run with calibration params
    reader = TxtinoutReader(path_to_txtinout)
    #tmp_path = reader.copy_swat(dir=None)
    #reader = TxtinoutReader(tmp_path)

    # Set up the SWAT model configuration
    reader.set_beginning_and_end_year(1990, 2012)
    reader.set_warmup(1)

    # Enable 'channel_sd' in the print.prt file (daily outputs enabled, others disabled)
    reader.enable_object_in_print_prt(obj='channel_sd', daily=True, monthly=False, yearly=False, avann=False)

    # Read plants.plt file
    filename = 'plants.plt'
    plants_reader = reader.register_file(filename, has_units=False, index='name')

    # Run SWAT with calibration parameters
    txt_in_out_result = reader.run_swat(calibration_params, show_output=False)

    # Register and read the channel_sd_day.txt results
    channel_sd = reader.register_file('channel_sd_day.txt', has_units=True)
    df = channel_sd.df.rename(columns={'yr': 'year', 'mon': 'month', 'day': 'day'})
    df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

    # Align simulated flow data with historical dates
    simulated_flow = df.set_index('date').reindex(historical_flow_data['date'])['flo_out'].fillna(0)
    observed_flow = historical_flow_data.set_index('date')['flow']

    # Calculate error
    error = calculate_error(simulated_flow, observed_flow)

    return error, {'calibration_path': reader.root_folder}
```

```python
#txtinout_folder = '../data/swatplus_rev60_demo/'
txtinout_folder = 'D:/Proyectos_GitHub/piura_river/data/swatplus_rev60_demo/'

swat_problem = SWATProblem(
    params={
        'plants.plt': ('name', [('bana', 'bm_e', 30, 50), ('bana', 'harv_idx', 0.3, 0.5)])
    },
    function_to_evaluate=function_to_minimize,
    param_arg_name='calibration_params',
    n_workers=1,
    parallelization='threads',
    debug=True,
    path_to_txtinout=txtinout_folder
)
```

```python
# Update the initial guess `x0` to match two parameters
x0 = denormalize(np.random.random(2), np.array([30, 0.3]), np.array([50, 0.5]))
n_simulations = 2

algorithm = CMAES(x0=x0, maxfevals=n_simulations)
termination = get_termination("n_eval", n_simulations)

# Run the optimization
result = minimize_pymoo(
    problem=swat_problem,
    algorithm=algorithm,
    termination=termination,
    seed=1,
    verbose=False,
    callback=None,
)

# Output the results
print("Best Parameters:", result.X)
print("Calibration Path:", result.pop.get("calibration_path"))
print("Error:", result.F)
```
