```python
import pandas as pd
import numpy as np
from pySWATPlus.TxtinoutReader import TxtinoutReader
from pySWATPlus.FileReader import FileReader
```

```python
#The folder pointed by txtinout_folder must have a .exe file to run the simulation
reader = TxtinoutReader(txtinout_folder)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load historical flow data globally if possible to avoid reloading
# Assuming 'historical_data.csv' has columns 'date' and 'flow'
historical_flow_data = pd.read_csv('D:/Proyectos_GitHub/piura_river/data/historical_flow_data.csv')
historical_flow_data['date'] = pd.to_datetime(historical_flow_data['date'])


# Assuming your DataFrame is named df
df = channel_sd.df
# Rename columns to ensure they match correctly
df = df.rename(columns={'yr': 'year', 'mon': 'month', 'day': 'day'})

# Now, create a date column using these correctly named columns
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

# Plot date vs. flo_out
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['flo_out'], label='Flow Out')
plt.xlabel('Date')
plt.ylabel('Flo Out')
plt.title('Date vs Flo Out')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the simulated flow data
df = channel_sd.df.rename(columns={'yr': 'year', 'mon': 'month', 'day': 'day'})
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df_filtered = df[(df['date'] >= '2004-01-01') & (df['date'] <= '2012-12-31')]

historical_flow_filtered = historical_flow_data[
    (historical_flow_data['date'] >= '2004-01-01') & (historical_flow_data['date'] <= '2012-12-31')
]

# Plot both filtered simulated and observed flows
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['date'], df_filtered['flo_out'], label='Simulated Flow (Flow Out)', color='blue')
plt.plot(historical_flow_filtered['date'], historical_flow_filtered['flow'], label='Observed Flow', color='orange')
plt.xlabel('Date')
plt.ylabel('Flow')
plt.title('Simulated vs Observed Flow (2004-2012)')
plt.legend()
plt.grid(True)
plt.show()
```
