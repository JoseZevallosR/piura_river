```python
import pandas as pd
import numpy as np
from pySWATPlus.TxtinoutReader import TxtinoutReader
from pySWATPlus.FileReader import FileReader
```

```python
txtinout_folder = '../data/swatplus_rev60_demo/'
filename =  'plants.plt'
```

```python
#The folder pointed by txtinout_folder must have a .exe file to run the simulation
reader = TxtinoutReader(txtinout_folder)
```

```python
#Read plants.plt file
plants_reader = reader.register_file(filename, 
                    has_units=False,
                    index='name')
```

```python
plants_reader.header_file
```

```python
plants_reader.df
```

```python
reader.set_beginning_and_end_year(2003, 2012)
```

```python
reader.set_warmup(1)
```

```python
#Enable the 'channel_sd' in the print.prt file (the other ones are mantained as they are)
reader.enable_object_in_print_prt(obj='channel_sd', daily=True, monthly=False, yearly=False, avann=False)
```

```python
#Run the model with the default settings
reader.run_swat(show_output=True)
```

```python
#Read the results
channel_sd = reader.register_file('channel_sd_day.txt', 
                    has_units=True)
channel_sd.df
```

```python
channel_sd.df.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
df = channel_sd.df
# Rename columns to ensure they match correctly
df = df.rename(columns={'yr': 'year', 'mon': 'month', 'day': 'day'})

# Now, create a date column using these correctly named columns
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

# Plot date vs. flo_out
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['flo_out'], label='Flow Out')
plt.xlabel('Date')
plt.ylabel('Flo Out')
plt.title('Date vs Flo Out')
plt.legend()
plt.grid(True)
plt.show()
```
